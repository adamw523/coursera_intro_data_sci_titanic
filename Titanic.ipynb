# Adding age / faire to decision

```python
# class
train = pd.read_csv('train.csv')
train['ac'] = train['age'].apply(lambda x: 'child' if x < 18 else 'adult')
train[['ac', 'sex', 'survived']].groupby(['sex', 'ac']).aggregate([np.size, np.sum, np.mean])
```

```python
train = pd.read_csv('train.csv')
train['age'].describe()

# split fares into 4 buckets
cuts = pd.qcut(train['fare'], 3)
# cuts.levels[0]
# cuts.labels
train['fare_level'] = cuts.labels
train['fare_level_values'] = train['fare_level'].apply(lambda x: cuts.levels[x])
train.head()
```

```python
cutdf = train[['sex', 'fare_level_values', 'pclass', 'survived']]
g = cutdf.groupby(['sex', 'pclass', 'fare_level_values'])
g.aggregate([np.size, np.sum, np.mean])
```

```python
test = pd.read_csv('test.csv')
```

```python
# did not survive if male or if female in 3rd class and paid more than 26 fare
test['survived'] = test.apply(lambda x: 0 if x['sex'] == 'male' else (0 if x['pclass'] == 3 and x['fare'] > 26 else 1), axis=1)
test[['survived']].to_csv('genderbasedfaremodelpandas.csv', index=False)
test.head()
# Kaggle score now is: 	 0.77512
```

# Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

sex_map = {'male': 1, 'female': 2}
embarked_map = {'S': 1, 'C': 2, 'Q': 3}

def fixvals(df):
    df['sex'] = df['sex'].apply(lambda x: sex_map.get(x, 0))
    df['embarked'] = df['embarked'].apply(lambda x: embarked_map.get(x, 0))
    df['age'] = df['age'].fillna(0)
    df['fare'] = df['fare'].fillna(0)
    return df

train = fixvals(train)
train = train[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']]

test = fixvals(test)
test = test[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']]

train.head()
```

```python
# Create the random forest object which will include all the parameters
# for the fit
rf = RandomForestClassifier(n_estimators = 100)

# Fit the training data to the training output and create the decision
# trees
rf = rf.fit(train[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']], train[['survived']])

# Take the same decision trees and run on the test data
output = rf.predict(test)
```

```python
output = pd.DataFrame(output, columns=['survived'])

#save to CSV
output[['survived']].to_csv('randomforestmodel.csv', index=False)
# Kaggle score: 0.74641
```
